## Remove Outlier using IQR

```python
df.shape
```

```python
df.head()
```

```python
#specify the column name to remove outlier from dataframe

column_name = ['Visibility']

for column_name in column_name:
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3-Q1
    df = df[~((df[column_name] <(Q1 - 1.5 * IQR)) | (df[column_name] > (Q3 + 1.5 * IQR)))]

df.head()    
```

## Split the Dataset Into Dependent and Independent Variable (Test Train Split) 

```python
X = df.drop('ProductOutletSales', axis=1)
y = df['ProductOutletSales']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state=0)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python
# Create a DecisionTreeRegressor object
dtree = DecisionTreeRegressor()

# Define the hyperparameters to tune and their values
param_grid = {
    'max_depth': [2, 4, 6, 8],
    'min_samples_split': [2, 4, 6, 8],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Create a GridSearchCV object
grid_search = GridSearchCV(dtree, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best parameters found: ", grid_search.best_params_)
```

```python
# Train the DecisionTreeRegressor with the best hyperparameters
best_params = grid_search.best_params_
dtree = DecisionTreeRegressor(**best_params)
dtree.fit(X_train, y_train)
```

```python
from sklearn import metrics
import math

# Train the model with the best hyperparameters on the training set
dtree_best = DecisionTreeRegressor(**best_params)
dtree_best.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dtree_best.predict(X_test)

# Calculate evaluation metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mape = metrics.mean_absolute_percentage_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)

# Print the evaluation metrics
print("Mean Absolute Error (MAE): {}".format(mae))
print("Mean Absolute Percentage Error (MAPE): {}".format(mape))
print("Mean Squared Error (MSE): {}".format(mse))
print("R-squared (R2): {}".format(r2))
print("Root Mean Squared Error (RMSE): {}".format(rmse))
```

```python
imp_df = pd.DataFrame({
    'Feature Name': X_train.columns,
    'Importance':dtree_best.feature_importances_
})

fi= imp_df.sort_values(by='Importance', ascending=False)

fi2 = fi.head(10)
plt.figure(figsize=(10,8))
sns.barplot(data=fi2, x="Importance", y="Feature Name")
plt.title("Feature Importance Each Attribute (Decision Tree Regressor)", fontsize=18)
plt.xlabel("Importance", fontsize=16)
plt.xlabel("Feature Name", fontsize=16)
plt.show()
```

## Insights:

+ **ProductMRP**: This feature has the longest bar, indicating it is the most important feature for predicting the target variable. In the context of sales analysis, this might mean that the Maximum Retail Price of the product is a significant factor in predicting sales.


+ **OutletType**: This feature is also highly important, suggesting that the type of outlet has a strong influence on sales predictions.


+ **Visibility**: This feature has a smaller bar compared to ProductMRP and OutletType, indicating it is less important but still plays a role in the model's predictions.


+ **Other Features**: Features like ProductType, OutletEstablishment_Year, FatContent, OutletSize, Weight, OutletId, and LocationType have relatively shorter bars, indicating they have less influence on the model's predictions compared to ProductMRP and OutletType.
